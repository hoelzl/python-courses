## More detailed description of variables

<img src="img/variables-01.svg" style="float:right;margin:auto;width:50%"/>

A *variable* is
- a <span style="color:red;">"reference"</span> to an "object"
- which has a <span style="color:red;">name</span>.

<span style="color:blue;">An object </span>
can be referenced by<span style="color:blue;"> several variables </span>!

<img src="img/variables-01.svg" style="float:right;margin:auto;width:50%"/>

A variable is
- generated by `name = value`
- read by `name`
- changed by `name=value`

Creating and changing variable <br/>
are *statements*.

```
length_birch_alley = 20
print(length_birch_alley)
length_birch_alley = 25
print(length_birch_alley)
```

## Properties of variables in Python

- A variable can store values of any data type
    - There are no `int` variables, etc.
    - Another way of saying this: Python is dynamically typed
- Variables must be created before they are used
- You can assign new values to variables
    - The *old value* of the variable can be used on the
      right hand side:<br/> `jobs = jobs + 1`

```
x = "Hallo!"
print(x)
x = 123
print(x)
x = x + 1
print(x)
x += 1
print(x)
```

## Mini-Workshop: Pirates

According to legend, the booty in pirate gangs was divided fairly amongst all
pirates. If the booty could not be shared fairly, the captain received the excess
share.

How many gold doubloons does each pirate in a gang of 8 (7 pirates + 1 captain) get
when looting a treasure with 1000 gold doubloons?

(Use variables to make the calculation clearer.)

```
anzahl_piraten = 8
beute_gesamt = 1000
beute_pro_pirat = beute_gesamt // anzahl_piraten
beute_pro_pirat
```

How many extra gold doubloons does the captain get?

```
# noinspection NonAsciiCharacters
beute_kapitän = beute_gesamt % anzahl_piraten
beute_kapitän
```
